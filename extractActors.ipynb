```
# Actors

def process(string):
  string = string.replace('{', '')
  string = string.replace('[', '')
  string = string.replace(']', '')
  string = string.replace('}', '')
  string = string.replace('<br/>', ',')
  string = string.split('|')
  if(len(string) != 1):
    string = string[1:]
  # print(string)
  ret = ""
  for word in string:
    ret += word + " "
  return ret

counter = 30

for infobox in infoboxes:
  # print(infobox.keys())
  if counter == 0:
    break
  try:
    keys = ["name",
      "image",
      "birth_name",
      "other_names",
      "birth_date",
      "birth_place",
      "death_date",
      "death_place",
      "nationality",
      "education",
      "occupation",
      "alma_mater",
      "years_active",
      "height",
      "known_for",
      "spouse",
      "awards",
      "children",
      "parents",
      "relatives"]
    hindi_keys = ["नाम",
      "चित्र",
      "जन्म नाम",
      "अन्य नाम",
      "जन्म दिन",
      "जन्म स्थान",
      "मौत की तिथि",
      "मृत्यु का स्थान",
      "राष्ट्रीयता",
      "शिक्षा",
      "व्यवसाय",
      "शिक्षा प्राप्त की",
      "सक्रिय वर्ष",
      "ऊंचाई",
      "प्रसिद्धि कारण",
      "जीवनसाथी",
      "पुरस्कार",
      "बच्चे",
      "माता-पिता",
      "संबंधी"]
    for i in range(len(keys)):
      if keys[i] in infobox and keys[i] != "birth_date":
        temp = infobox[keys[i]]
        key = hindi_keys[i]
        # print(temp)
        temp = process(temp)
        if keys[i] != "image":
          translation = translator.translate(temp, dest = "hi")
          print(key, ": ", translation.text)
        else:
          print(key, ": ", temp)
  except:
    print("No data found for the movie")
  print()
  counter -= 1
    
# translation = translator.translate("net_worth", dest = "hi")
# print(translation.text)
```
